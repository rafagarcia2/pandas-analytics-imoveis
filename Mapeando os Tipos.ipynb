## Seleções e Frequências

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados
```

```python
# Selecao somente os imóveis classificados com tipo 'Apartamento'
selecao = dados['Tipo'] == 'Apartamento'
selecao
```

```python
n_apart = dados[selecao].shape[0]
n_apart
```

```python
# Selecao dos imóveis classificados como qualquer tipo de 'Casa'
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n_casas = dados[selecao].shape[0]
n_casas
```

```python
# Selecao dos imóveis com área entre 60 e 100 metros quadrados, incluindo os limites
# <= Area <= 100
selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n_area = dados[selecao].shape[0]
n_area
```

```python
# Selecao dos imóveis com menos de 4 quartos e aluguel menor que R$ 2.000,00
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n_quartos = dados[selecao].shape[0]
n_quartos
```

```python
print("Nº de imóveis classificados com tipo \'Apartamento\': " + str(n_apart))
print("Nº de imóveis classificados como qualquer tipo de \'Casa\': " + str(n_casas))
print("Nº de imóveis com área entre 60 e 100 metros quadrados: " + str(n_casas))
print("Nº de imóveis com menos de 4 quartos e aluguel menor que R$ 2.000,00: " + str(n_quartos))
```
